```python
# Task 2 (15 marks) For this task, you are to write code that computes the value of acceleration of object, given the values of displacement, time, and initial velocity.

Instructions
Your program should ask for user inputs to get the values of displacement (how much distance the object has travelled), time, and the initial velocity of the object. It should then calculate the acceleration of the object and print this value out.

Here is a flowchart which illustrates the expected program flow:

acceleration.png

The equation to compute the displacement of the object, s, is given by:
s=a∗t22+u∗t
where s is the displacement of the object,a is the acceleration, t is the time, and u is the initial velocity.

Requirements
To achieve full marks for this task, you must follow the instructions above when writing your solution. Additionally, your solution must adhere to the following requirements:

Appropriate data types must be used for all user inputs.
The output must be printed with the value between the description and the unit of measurement, as shown in the example runs.
```

```python
# Write your solution here
a=float(input("Acceleration (m/s^2) : "))
u=float(input("Initial velocity (m/s):  "))
t=float(input("Time (s): "))
s=(float(a*t**2/2 + u*t))
print("Object displacement " + str(s) + ' m')
```

```python
def calculate_body_mass_index(height,weight):
     return (weight/(height**2))


# Write your solution here
w = float(input("Weight (kg) : "))
h = float(input("Height (m) : "))

bmi = calculate_body_mass_index(h,w)
if (bmi <= 18.5):
    print("Your BMI result is : Underweight")
elif (18.5 < bmi <= 24.9):
    print("Your BMI result is : Healthy Weight")
elif (25<bmi<=29.29):
    print("Your BMI result is : Overweight")
else:
    print("Your BMI result is : Obese")
```
